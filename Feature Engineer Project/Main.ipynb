```python
%matplotlib inline
sns.boxplot(x=numeric_df['SalePrice'])
```

```python
def outlier_treatment(col):
    sorted(col)
    Q1, Q3 = np.percentile(col, [25, 75])
    IQR = Q3 - Q1
    lower_range = Q1 - (1.5 * IQR)
    upper_range = Q3 + (1.5 * IQR)
    return lower_range, upper_range
```

```python
lower_range, upper_range = outlier_treatment(numeric_df['SalePrice'])
print("Lower Range:",lower_range)
print("Upper Range:",upper_range)
```

```python
lower_numeric_df = numeric_df[numeric_df['SalePrice'].values < lower_range]
lower_numeric_df
```

```python
upper_numeric_df = numeric_df[numeric_df['SalePrice'].values > upper_range]
upper_numeric_df
```

```python
upper_outliers = upper_numeric_df.value_counts().sum(axis=0)
print("Outliers: ",upper_outliers)
```

```python
upper_index = list(upper_numeric_df[ upper_numeric_df['SalePrice'] > upper_range ].index)
print("Index: ",upper_index)
```

```python
print("Shape Before Dropping Outlier Rows:", numeric_df.shape)
numeric_df.drop(upper_index, inplace = True)
print("Shape After Dropping Outlier Rows:", numeric_df.shape)
```


#### We have successfully dropped the unwanted rows
